##DFT

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import ifft, fft, fftfreq
```

```python
data = np.genfromtxt('monthrg.dat')
year = data[:,0]
month = data[:,1] 
day = data[:,2]
avera = data[:,3]

ano = []
mes = []
prom = []
```

```python
for i in range(len(day)): 
    if(day[i]!=0):
        ano.append(year[i])
        mes.append(month[i])
        prom.append(avera[i])

ano = [ano[i] + mes[i]/12 for i in range(len(ano))]

plt.plot(ano,prom)
```

Se puede ver que entre 1650 y 1725 el ciclo solar fue bastante bajo y un poco fuera de lo normal.

```python
T=len(prom)
fouTranProm = fft(prom)
omega = fftfreq(T, 1./12)
```

```python
h = max(np.abs(fouTranProm)[omega>0.01])

n=0
for i in range(len(fouTranProm[omega>0.01])):
    if(np.abs(fouTranProm[omega>0.01][i])==h):
        n = i
        
print "El periodo Solar de es",1/omega[n], "años aproximadamente."
```

```python
inverse = ifft(fouTranProm)
```

```python
plt.figure(figsize=(15,10))
plt.plot(ano[1988:],prom[1988:],'b',alpha=0.9)
plt.plot(ano[1988:],np.abs(inverse[1988:]),'r',linewidth=1.)
plt.title('Promedio de Manchas Solares Mensuales')
plt.xlabel('Anos')
plt.ylabel('Manchas Solares/Mes')
plt.show()
```
