```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = np.array([0, np.pi/2,np.pi, 3*np.pi/2])
y = np.sin(x)
inter = interpolate.splrep(x, y, s=0)
xnew = np.array([0, np.pi/2,np.pi, 3*np.pi/2])
ynew = interpolate.splev(xnew, inter, der=0)
```

```python
plt.figure()
plt.plot(x, y, 'x', xnew, ynew, xnew, np.sin(xnew), x, y, 'b')
plt.show()
```
