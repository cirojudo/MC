```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
Nt=2000
Nl=50
L=10. # Lado del cubo
tmin=0
tmax=.3

xmin=-L/2
xmax=L/2
ymin=-L/2
ymax=L/2

dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
dy=(ymax-ymin)/Nl

v=425. # m/s

A=1./np.sqrt(2)
```

```python
drumCara1=np.zeros((Nt,Nl+2,Nl+2))
drumCara2=np.zeros((Nt,Nl+2,Nl+2))
drumCara3=np.zeros((Nt,Nl+2,Nl+2))
drumCara4=np.zeros((Nt,Nl+2,Nl+2))
drumCara5=np.zeros((Nt,Nl+2,Nl+2))
drumCara6=np.zeros((Nt,Nl+2,Nl+2))
```

```python

xcoords_c1=np.linspace(xmin,xmax,Nl)
ycoords_c1=np.linspace(xmin,xmax,Nl)

xcoords_c2=np.linspace(xmin,xmax,Nl)
ycoords_c2=np.linspace(xmin,xmax,Nl)
xmesh_2,ymesh_2=np.meshgrid(xcoords_c2,ycoords_c2)

xcoords_c3=np.linspace(xmin,xmax,Nl)
ycoords_c3=np.linspace(xmin,xmax,Nl)

xcoords_c4=np.linspace(xmin,xmax,Nl)
ycoords_c4=np.linspace(xmin,xmax,Nl)
xmesh_4,ymesh_4=np.meshgrid(xcoords_c4,ycoords_c4)

xcoords_c5=np.linspace(xmin,xmax,Nl)
ycoords_c5=np.linspace(xmin,xmax,Nl)

xcoords_c6=np.linspace(xmin,xmax,Nl)
ycoords_c6=np.linspace(xmin,xmax,Nl)
xmesh_6,ymesh_6=np.meshgrid(xcoords_c6,ycoords_c6)
```

```python
drumCara2[0,1:-1,1:-1] = np.exp(-xmesh_2**2/(2*A**2))*np.exp(-ymesh_2**2/(2*A**2))
drumCara2[1,1:-1,1:-1] = drumCara2[0,1:-1,1:-1]+((v*dt/dx)**2)/2.*(np.roll(drumCara3[0,1:-1,1:-1],1,axis=0)
                                                                 +np.roll(drumCara2[0,1:-1,1:-1],-1,axis=0)
                                                                 +np.roll(drumCara2[0,1:-1,1:-1],1,axis=1)
                                                                 +np.roll(drumCara2[0,1:-1,1:-1],-1,axis=1)
                                                                 -4*drumCara2[0,1:-1,1:-1])

drumCara4[0,1:-1,1:-1] = np.exp(-xmesh_4**2/(2*A**2))*np.exp(-ymesh_4**2/(2*A**2))
drumCara4[1,1:-1,1:-1] = drumCara4[0,1:-1,1:-1]+((v*dt/dx)**2)/2.*(np.roll(drumCara4[0,1:-1,1:-1],1,axis=0)
                                                                 +np.roll(drumCara4[0,1:-1,1:-1],-1,axis=0)
                                                                 +np.roll(drumCara4[0,1:-1,1:-1],1,axis=1)
                                                                 +np.roll(drumCara4[0,1:-1,1:-1],-1,axis=1)
                                                                 -4*drumCara4[0,1:-1,1:-1])
```

```python
for i in range(2,Nt):
    cubo1 = drumCara1[i-1]
    cube1 = drumCara1[i-2]
    cubo2 = drumCara2[i-1]
    cube2 = drumCara2[i-2]
    cubo3 = drumCara3[i-1]
    cube3 = drumCara3[i-2]
    cubo4 = drumCara4[i-1]
    cube4 = drumCara4[i-2]
    cubo5 = drumCara5[i-1]
    cube5 = drumCara5[i-2]
    cubo6 = drumCara6[i-1]
    cube6 = drumCara6[i-2]
```

```python
    cubo1[1:-1,-1] = cubo2[1:-1,1]
    cubo1[1:-1,0] = cubo4[1:-1,-2]
    cubo1[0,1:-1] = cubo5[1:-1,1]
    cubo1[-1,1:-1] = cubo6[1:-1,1]
    
    cubo2[1:-1,0] = cubo1[1:-1,-2]
    cubo2[1:-1,-1] = cubo3[1:-1,1]
    cubo2[0,1:-1] = cubo5[-2,1:-1]
    cubo2[-1,1:-1] = cubo6[1,1:-1]
    
    cubo3[1:-1,0] = cubo2[1:-1,-2]
    cubo3[1:-1,-1] = cubo4[1:-1,1]
    cubo3[0,1:-1] = cubo5[1:-1,-2]
    cubo3[-1,1:-1] = cubo6[1:-1,-2]
    
    cubo4[1:-1,-1] = cubo1[1:-1,1]
    cubo4[1:-1,0] = cubo3[1:-1,-2]
    cubo4[0,1:-1] = cubo5[1,1:-1]
    cubo4[-1,1:-1] = cubo6[-2,1:-1]
    
    cubo5[1:-1,0] = cubo1[1,1:-1]
    cubo5[-1,1:-1] = cubo2[1,1:-1]
    cubo5[1:-1,-1] = cubo3[1,1:-1]
    cubo5[0,1:-1] = cubo4[1,1:-1]
    
    cubo6[1:-1,0] = cubo1[-2,1:-1]
    cubo6[0,1:-1] = cubo2[-2,1:-1]
    cubo6[1:-1,-1] = cubo3[-2,1:-1]
    cubo6[-1,1:-1] = cubo4[-2,1:-1]
        
    cube1[0,1:-1] = cube5[1:-1,1]
    cube1[1:-1,0] = cube4[1:-1,-2]
    cube1[-1,1:-1] = cube6[1:-1,1]
    cube1[1:-1,-1] = cube2[1:-1,1]
    
    cube2[0,1:-1] = cube5[-2,1:-1]
    cube2[1:-1,0] = cube1[1:-1,-2]
    cube2[-1,1:-1] = cube6[1,1:-1]
    cube2[1:-1,-1] = cube3[1:-1,1]
    
    cube3[0,1:-1] = cube5[1:-1,-2]
    cube3[1:-1,0] = cube2[1:-1,-2]
    cube3[-1,1:-1] = cube6[1:-1,-2]
    cube3[1:-1,-1] = cube4[1:-1,1]
    
    cube4[0,1:-1] = cube5[1,1:-1]
    cube4[1:-1,0] = cube3[1:-1,-2]
    cube4[-1,1:-1] = cube6[-2,1:-1]
    cube4[1:-1,-1] = cube1[1:-1,1]
    
    cube5[0,1:-1] = cube4[1,1:-1]
    cube5[1:-1,0] = cube1[1,1:-1]
    cube5[-1,1:-1] = cube2[1,1:-1]
    cube5[1:-1,-1] = cube3[1,1:-1]
    
    cube6[0,1:-1] = cube2[-2,1:-1]
    cube6[1:-1,0] = cube1[-2,1:-1]
    cube6[-1,1:-1] = cube4[-2,1:-1]
    cube6[1:-1,-1] = cube3[-2,1:-1] 
```

```python
for i in range(2,Nt):
    drumCara1[i]=((np.roll(cubo1,1,axis=0)+np.roll(cubo1,-1,axis=0)+np.roll(cubo1,1,axis=1)+np.roll(cubo1,-1,axis=1))*(v*dt/dx)**2)+2*(1.-2.*(v*dt/dx)**2)*cubo1-cube1
    
    drumCara2[i]=((np.roll(cubo2,1,axis=0)+np.roll(cubo2,-1,axis=0)+np.roll(cubo2,1,axis=1)+np.roll(cubo2,-1,axis=1))*(v*dt/dx)**2)+2*(1.-2.*(v*dt/dx)**2)*cubo2-cube2
    
    drumCara3[i]=((np.roll(cubo3,1,axis=0)+np.roll(cubo3,-1,axis=0)+np.roll(cubo3,1,axis=1)+np.roll(cubo3,-1,axis=1))*(v*dt/dx)**2)+2*(1.-2.*(v*dt/dx)**2)*cubo3-cube3
    
    drumCara4[i]=((np.roll(cubo4,1,axis=0)+np.roll(cubo4,-1,axis=0)+np.roll(cubo4,1,axis=1)+np.roll(cubo4,-1,axis=1))*(v*dt/dx)**2)+2*(1.-2.*(v*dt/dx)**2)*cubo4-cube4
    
    drumCara5[i]=((np.roll(cubo5,1,axis=0)+np.roll(cubo5,-1,axis=0)+np.roll(cubo5,1,axis=1)+np.roll(cubo5,-1,axis=1))*(v*dt/dx)**2)+2*(1.-2.*(v*dt/dx)**2)*cubo5-cube5
    
    drumCara6[i]=((np.roll(cubo6,1,axis=0)+np.roll(cubo6,-1,axis=0)+np.roll(cubo6,1,axis=1)+np.roll(cubo6,-1,axis=1))*(v*dt/dx)**2)+2*(1.-2.*(v*dt/dx)**2)*cubo6-cube6
```

```python
%matplotlib

figure = plt.figure(figsize=(20,15))
ex = np.linspace(-2*L,2*L,4*Nl)
ey = np.linspace(-3*L/2,3*L/2,3*Nl)
Total = np.zeros((Nt,3*Nl,4*Nl)) 

for i in range(len(Total)):
    
    Total[i,Nl:2*Nl,0:Nl] = drumCara1[i,1:-1,1:-1]
    Total[i,Nl:2*Nl,Nl:2*Nl] = drumCara2[i,1:-1,1:-1]
    Total[i,Nl:2*Nl,2*Nl:3*Nl] = drumCara3[i,1:-1,1:-1]
    Total[i,Nl:2*Nl,3*Nl:4*Nl] = drumCara4[i,1:-1,1:-1]
    Total[i,2*Nl:3*Nl,Nl:2*Nl] = drumCara5[i,1:-1,1:-1]
    Total[i,0:Nl,Nl:2*Nl] = drumCara6[i,1:-1,1:-1]

plt.pcolor(ex,ey,Total[0],cmap='gist_heat',vmin=0.,vmax=1.)

plt.xlim(-2*L + dx,2*L - dx)
plt.ylim(-3*L/2 + dy,3*L/2 -dy)

def animate(i):  
    plt.pcolor(ex,ey,Total[i],cmap='gist_heat',vmin=0.,vmax=1.)
    plt.xlim(-2*L + dx,2*L - dx)
    plt.ylim(-3*L/2 + dy,3*L/2 -dy)
    
animacion = animation.FuncAnimation(figure, animate, np.arange(0,len(drumCara1),100), interval=1, blit=False,repeat_delay=100)

plt.show()
```
