#Punto 1.a

Programe una función de python llamada lagrange que reciba el número N de puntos $\{t_i , f_i \}$ para $i = 0..., \ N − 1$, y que entregue de regreso una expresión simbólica para el polinomio de grado $N − 1$ que los interpola. Puede usar var para definir símbolos de SymPy de forma programática y eval para convertir strings en nombres de objetos. Abajo a la izquierda se muestra el output en algunos casos.

```python
def lagrange(N): #inicializamos la función lagrange
    n = 1 #numero de iteraciones u orden
    t = var('t') #tiempo
    orden = 0 #orden de la función
    f_i = symarray('f',N) #se inicializan los f_i
    t_i = symarray('t',N) #se inicializan los t_i
    
    for i in range(N):
         
        for j in range(N):
            if(i!=j):
                n *=(t - t_i[j])/(t_i[i] - t_i[j]) #se realiza la multiplicatoria
        orden += f_i[i]*n 
        
    return orden #se imprime la respuesta de lagrange
```

```python
lagrange(1)
```

```python
lagrange(2)
```

```python
lagrange(3)
```

##Punto 1.b

Ahora, en el contexto de los métodos multistep de Adams-Bashforth defina una función llamada adams que reciba el orden m a considerar y de regreso entregue el incremento correspondiente. Para ello tome $t_i = t_n − i ∗ h$ y $f_i = f(t_i )$ siendo $h$ el timestep. Para ello debe utilizar la función definida en el anterior literal, y utilizar integrate y simplify de SymPy. Arriba a la derecha se muestra el output en algunos casos.

```python
def adams(m):#inizializamoa la función de adams
    
    lag = lagrange(m) #tomamos la función de lagrange hecha anteriorimente
    h_i= symbols('h') #inicializamos los h_i
    t_i= symarray('t',m) #inicializamos los t_i
    t = symbols('t') #inicializamos el t en 0
    
    for k in range (m): #recorrido sobre la integración del polinomio
        lag = lag.subs(t_i[k],t_i[0]-k*h_i)
        orden=simplify(integrate(lag,(t,t_i[0],t_i[0]+h_i)))
        
    return orden #se retorna el orden de la función 
```

```python
adams(2)
```

```python
adams(3)
```

##Punto 2

```python
xyzMin = -1500.
xyzMax = 1500.
Cons=101
xyz = np.linspace(xyzMin,xyzMax,Cons)
# definicion de las coordenadas 

xv, yv, zv = np.meshgrid(xyz, xyz, xyz) #las cordenadas se vectorizan
#CONSTANTES  

h=30. 
G=6.67384*10**(-11.) # Nm2/kg2 
density=400.  # kg/m3
e= G*density*(100**2)*270.**3./1500. #correcion de los bordes
```

###a)

```python
# Como el punto (0,0,0) esta incluido, se suma 3 para no tener una indeterminacion 
r=(np.sqrt(xv**2. + yv**2. + zv**2.)+3 ) 
Gm=(G*density*270.**3)
po = -Gm/r + (e/100**2)
#condiciones de frontera del enunciado, en los bordes es 0
po[0] = 0.
po[-1] = 0.
po[:,0] = 0.
po[:,-1] = 0.
po[:,:,0] = 0.
po[:,:,-1] = 0.

itera = 1000 # Número de iteraciones para resolver la ecuación de Poisson para el potencial gravitacional.

#esto fue tomado del cuaderno de ejemplos de la clase
for i in range(itera):
    po = 1/6.*(np.roll(po,1,axis=0) + np.roll(po,-1,axis=0) + np.roll(po,1,axis=1) + np.roll(po,-1,axis=1) + np.roll(po,1,axis=2) + np.roll(po,-1,axis=2)) 
    
    po[46:55,46:55,46:55] -= (1./6.)*4.*np.pi*G*density*h**2.
    #condiciones de frontera en cada iteracion 
    po[0] = 0.
    po[-1] = 0.
    po[:,0] = 0.
    po[:,-1] = 0.
    po[:,:,0] = 0.
    po[:,:,-1] = 0. 
```

```python
#bloque de codigo para generar la grafica
plt.figure(figsize=(15,12))

plt.pcolor(xyz,xyz,po[50]*100**2,cmap='gist_heat') # Se hace la conversion de unidades
plt.axis('equal')
plt.colorbar() 
CS=plt.contour(xyz,xyz,po[50]*100**2,[-16.,-8.,-4.,-2.,-1.],linestyles = 'dashed',colors = 'k') 
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel(r'$y/m$',fontsize=20)
plt.ylabel(r'$z/m$',fontsize=20)
plt.title(r'$\phi/cm^2s^{-2}$',fontsize=20)

plt.show()
```

```python
po1 = np.zeros((Cons,Cons,Cons)) #se define otro para poder comparar resultados

xyzCube = np.linspace(-120.,120.,9)
#se suma el potencial generado por cada uno de los cubos pequeños
for i in range(len(xyzCube)):
    for j in range(len(xyzCube)):
        for k in range(len(xyzCube)):
            po1 += (-(G*density*270.**3.)/729.)/(np.sqrt((xv-xyzCube[i])**2. + (yv-xyzCube[j])**2. + (zv-xyzCube[k])**2.) +3.)
            
plt.figure(figsize=(15,10))

#correcion en los bordes, se le suma una constante
plt.plot(xyz,po1[50,50,:]*100**2+ e,label = 'sum + $\epsilon$') 
plt.scatter(xyz,po[50,50,:]*100**2,facecolors='none', label = 'Poisson') 
plt.xlabel(r'$z/m$',fontsize=20)
plt.ylabel(r'$\phi/cm^2s^{-1}$',fontsize=20)
plt.title(r'Potencial Gravitacional',fontsize=20)
plt.legend(loc = 'best')
plt.show()
```

```python
#tomado de los ejemplos de clase
poAtras = np.roll(po[50,50,:],-1) 
poAdelante = np.roll(po[50,50,:],1) 
dpo = (poAtras-poAdelante)/(2.*h) 
dpo = dpo[1:-1] # Se ignoran los datos de los extremos ya que no tienen sentido.

plt.figure(figsize=(15,10))

plt.scatter(xyz[1:-1],-dpo*10**6,facecolors='none') #se hace el cambio de unidades

plt.xlabel(r'$z/m$',fontsize=20)
plt.ylabel(r'$g/\mu m^2s^{-1}$',fontsize=20)
plt.title(r'Campo Gravitacional',fontsize=20)

plt.show()
```
